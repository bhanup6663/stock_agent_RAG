```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the tesla data
tesla_data = pd.read_csv('datasets/TSLA.csv')
```

```python
tesla_data.columns
```

```python
tesla_data['name'] = 'Tesla'
tesla_data['source'] = 'Kaggle'
```

```python
tesla_data.head
```

```python
#missing values
missing_values = tesla_data.isnull().sum()
print(missing_values)
```

```python
summary_stats = tesla_data.describe()
print(summary_stats)
```

```python
data_types = tesla_data.dtypes
print(data_types)
```

```python
duplicates = tesla_data.duplicated().sum()
print(duplicates)
```

```python
sns.boxplot(data=tesla_data[['Open', 'High', 'Low', 'Close', 'Adj Close', 'Volume']])
plt.show()
```

```python
tesla_data[['Open', 'High', 'Low', 'Close']].plot(figsize=(10, 6))
plt.title('Stock Prices Over Time')
plt.xlabel('Date')
plt.ylabel('Price')
plt.show()
```

```python
tesla_data['Volume'].plot(kind='bar', figsize=(10, 6))
plt.title('Trading Volume Over Time')
plt.xlabel('Date')
plt.ylabel('Volume')
plt.show()
```

```python
numeric_data = tesla_data.select_dtypes(include=['float64', 'int64'])
```

```python
sns.heatmap(numeric_data.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
tesla_data['Close'].rolling(window=20).mean().plot(figsize=(10, 6))
plt.title('20-Day Moving Average of Close Price')
plt.xlabel('Date')
plt.ylabel('Price')
plt.show()
```
