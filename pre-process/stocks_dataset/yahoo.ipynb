```python
data_yahoo = yf.download(tickers, start="2015-01-01", end="2023-01-01", group_by='ticker')
```

```python
stacked_data = data_yahoo.stack(level=0).reset_index()

# Rename the columns to lowercase, consistent with previous renaming
stacked_data.columns = ['date', 'Name', 'open', 'high', 'low', 'close', 'adj_close', 'volume']
```

```python
stacked_data['source'] = 'Yahoo Finance'
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.boxplot(data=stacked_data[['open', 'high', 'low', 'close', 'volume']])
plt.show()
```

Line Plot: To visualize the trend of stock prices over time.

```python
stacked_data[['open', 'high', 'low', 'close', 'volume']].plot(figsize=(10, 6))
plt.title('Stock Prices Over Time')
plt.xlabel('Date')
plt.ylabel('Price')
plt.show()
```

Correlation Heatmap: To understand the correlation between different columns.

```python
numeric_columns = stacked_data[['open', 'high', 'low', 'close', 'volume']]
sns.heatmap(numeric_columns.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

Moving Averages: To smooth out the data and identify trends

```python
stacked_data['close'].rolling(window=20).mean().plot(figsize=(10, 6))
plt.title('20-Day Moving Average of Close Price')
plt.xlabel('Date')
plt.ylabel('Price')
plt.show()
```
