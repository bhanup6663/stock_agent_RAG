```python
def load_sp500(filepath):
    sp500_data = pd.read_csv(filepath)
    sp500_data['source'] = 'SP500'
    sp500_data_cleaned = sp500_data.dropna()
    return sp500_data_cleaned
```

```python
def load_tesla(filepath):
    tesla_data = pd.read_csv(filepath)
    tesla_data['name'] = 'Tesla'
    tesla_data['source'] = 'Kaggle'
    return tesla_data
```

```python
import numpy as np
from sklearn.preprocessing import MinMaxScaler
```

```python
# Step 1: Handling erroneous values in 'adj_close'
# Replace negative or extreme 'adj_close' values with NaN
combined_data['adj_close'] = combined_data['adj_close'].apply(lambda x: np.nan if x < 0 else x)
```

```python
# Step 2: Handling missing values
# Check for missing values
missing_values = combined_data.isnull().sum()
print("Missing values before imputation:\n", missing_values)
```

```python
# Impute missing values for numeric columns using forward fill method (suitable for time series data)
combined_data.fillna(method='ffill', inplace=True)
```

```python
# Step 3: Removing or scaling outliers (focus on 'volume')
# Capping extreme values in 'volume' at the 99th percentile
volume_threshold = combined_data['volume'].quantile(0.99)
combined_data['volume'] = np.where(combined_data['volume'] > volume_threshold, volume_threshold, combined_data['volume'])
```

```python
# Step 4: Normalizing data using Min-Max Scaling
scaler = MinMaxScaler()
```

```python
# Apply scaling to numeric columns: 'open', 'high', 'low', 'close', 'adj_close', 'volume'
numerical_cols = ['open', 'high', 'low', 'close', 'adj_close', 'volume']
combined_data[numerical_cols] = scaler.fit_transform(combined_data[numerical_cols])
```
